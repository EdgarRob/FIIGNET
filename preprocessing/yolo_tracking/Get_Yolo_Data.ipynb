```python
import os
import shutil

# Define the source folder containing the 20 folders
source_folder = '/home/ugrad/serius/edgarrobitaille/AquaFinity-main/Deepfish'

# Define the destination folder
destination_folder = '/home/ugrad/serius/edgarrobitaille/AquaFinity-main'

# Create the "tr" and "te" folders in the destination folder
tr_folder = os.path.join(destination_folder, 'train')
te_folder = os.path.join(destination_folder, 'test')
os.makedirs(tr_folder, exist_ok=True)
os.makedirs(te_folder, exist_ok=True)

 # Create the corresponding "images" and "labels" folders in the "tr" folder
tr_images_folder = os.path.join(tr_folder, 'images')
tr_labels_folder = os.path.join(tr_folder, 'labels')
os.makedirs(tr_images_folder, exist_ok=True)
os.makedirs(tr_labels_folder, exist_ok=True)

# Create the corresponding "images" and "labels" folders in the "te" folder
te_images_folder = os.path.join(te_folder, 'images')
te_labels_folder = os.path.join(te_folder, 'labels')
os.makedirs(te_images_folder, exist_ok=True)
os.makedirs(te_labels_folder, exist_ok=True)

# Iterate through each of the 20 folders
for folder_name in os.listdir(source_folder):
    folder_path = os.path.join(source_folder, folder_name)
    
    # Check if the item is a directory
    if os.path.isdir(folder_path):
       
        
        # Process the "train" folder
        train_folder = os.path.join(folder_path, 'train')
        for file_name in os.listdir(train_folder):
            file_path = os.path.join(train_folder, file_name)
            if file_name.endswith('.jpg'):
                # Move the jpg file to the "tr/images" folder
                destination_path = os.path.join(tr_images_folder, file_name)
                shutil.copyfile(file_path, destination_path)
            elif file_name.endswith('.txt'):
                # Move the txt file to the "tr/labels" folder
                destination_path = os.path.join(tr_labels_folder, file_name)
                shutil.copyfile(file_path, destination_path)
        
        # Process the "valid" folder
        valid_folder = os.path.join(folder_path, 'valid')
        for file_name in os.listdir(valid_folder):
            file_path = os.path.join(valid_folder, file_name)
            if file_name.endswith('.jpg'):
                # Move the jpg file to the "te/images" folder
                destination_path = os.path.join(te_images_folder, file_name)
                shutil.copyfile(file_path, destination_path)
            elif file_name.endswith('.txt'):
                # Move the txt file to the "te/labels" folder
                destination_path = os.path.join(te_labels_folder, file_name)
                shutil.copyfile(file_path, destination_path)
```
